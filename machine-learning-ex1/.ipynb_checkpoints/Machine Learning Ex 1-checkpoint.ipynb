```python
import pandas as pd

data = pd.read_csv("./ex1data1.txt", header=None)
population = data[0].values
profits = data[1].values
```

```python
import matplotlib.pyplot as plt
import numpy as np

plt.scatter(population, profits, marker='+', color='red')
plt.xlabel('Population of City in 10,000s')
plt.ylabel('Profit in $10,000s')
```

```python
def compute_cost(X, y, theta):
    m = len(y)
    J = 1/(2*m) * np.sum(np.square(np.dot(X, theta) - y))
    return J
```

```python
theta = np.zeros(2)
X = np.array([[1, p] for p in population])
y = profits

compute_cost(X, y, theta)
```

```python
def gradient_descent(X, y, theta, alpha, iterations):
    m = len(y)
    for i in range(iterations):
        theta0 = theta[0] - alpha * (1/m) * np.sum(np.dot(X, theta) - y)
        theta1 = theta[1] - alpha * (1/m) * np.sum((np.dot(X, theta) - y) * X[:,1])
        theta[0] = theta0
        theta[1] = theta1
```

```python
theta = np.zeros(2)
iterations = 1500
alpha = 0.01

gradient_descent(X, profits, theta, alpha, iterations)
print(theta)
```

```python
plt.scatter(population, profits, marker='+', color='red')
plt.xlabel('Population of City in 10,000s')
plt.ylabel('Profit in $10,000s')
samples = np.linspace(min(population), max(population))
plt.plot(samples, theta[0] + theta[1] * samples)
```

```python
from mpl_toolkits.mplot3d import Axes3D
from matplotlib import cm

t0 = np.linspace(-10, 10, 20)
t1 = np.linspace(-1, 4, 20)

Xs, Ys = np.meshgrid(t0, t1)
Zs = np.array([compute_cost(X, y, [t0, t1]) for t0, t1 in zip(np.ravel(Xs), np.ravel(Ys))])
fig = plt.figure()
ax = fig.gca(projection="3d")
ax.view_init(azim=235)
Z = Zs.reshape(Xs.shape)
ax.plot_surface(Xs, Ys, Z, cmap=cm.rainbow)
```

```python
fig = plt.figure()
ax = fig.gca()
ax.plot(theta[0], theta[1], 'r*')
space = np.logspace(-2, 3, 20)
CS = plt.contour(Xs, Ys, Z, space)
plt.show()
```
